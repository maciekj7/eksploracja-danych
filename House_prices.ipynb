```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="pastel", font_scale=1.1)
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv("train.csv")

print("ðŸ”¹ Rozmiar danych:", df.shape)
print("\nðŸ”¹ Pierwsze 5 wierszy:")
display(df.head())

print("\nðŸ”¹ Typy danych:")
display(df.dtypes.value_counts())

print("\nðŸ”¹ Podstawowe statystyki opisowe (dla zmiennych numerycznych):")
display(df.describe())

target = 'SalePrice'

print("\nPodstawowe statystyki zmiennej docelowej:")
print(df[target].describe())
print("\nMediana ceny:", df[target].median())
print("Åšrednia cena:", df[target].mean())

plt.figure(figsize=(10,6))
sns.histplot(df[target], bins=30, kde=True)
plt.title("RozkÅ‚ad cen domÃ³w (SalePrice)")
plt.xlabel("Cena")
plt.ylabel("Liczba domÃ³w")
plt.show()

plt.figure(figsize=(10,4))
sns.boxplot(x=df[target])
plt.title("Boxplot cen domÃ³w")
plt.xlabel("Cena")
plt.show()

skew_val = df[target].skew()
print(f"SkoÅ›noÅ›Ä‡ rozkÅ‚adu cen: {skew_val:.2f}")
if skew_val > 1:
    print("RozkÅ‚ad jest prawoskoÅ›ny (duÅ¼o tanich domÃ³w, maÅ‚o bardzo drogich).")
elif skew_val < -1:
    print("RozkÅ‚ad jest lewoskoÅ›ny (duÅ¼o drogich domÃ³w, maÅ‚o tanich).")
else:
    print("RozkÅ‚ad jest zbliÅ¼ony do symetrycznego.")

area_cols = [col for col in df.columns if ('Area' in col) or ('SF' in col)]
print("\nKolumny zwiÄ…zane z powierzchniÄ…:")
print(area_cols)

corrs = df[area_cols + [target]].corr()[target].sort_values(ascending=False)
print("\nKorelacja powierzchni z cenÄ…:")
display(corrs)

plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, alpha=0.6)
plt.title("ZaleÅ¼noÅ›Ä‡: powierzchnia mieszkalna (GrLivArea) vs cena")
plt.xlabel("Powierzchnia mieszkalna (stopy kw.)")
plt.ylabel("Cena domu")
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='LotArea', y='SalePrice', data=df, alpha=0.6, color='orange')
plt.title("ZaleÅ¼noÅ›Ä‡: powierzchnia dziaÅ‚ki (LotArea) vs cena")
plt.xlabel("Powierzchnia dziaÅ‚ki")
plt.ylabel("Cena domu")
plt.show()

selected_cols = ['GrLivArea', 'TotalBsmtSF', 'LotArea']

plt.figure(figsize=(10,6))
for col in selected_cols:
    sns.regplot(x=df[col], y=df['SalePrice'], label=col, scatter_kws={'alpha':0.4})
plt.legend()
plt.title("PorÃ³wnanie wpÅ‚ywu rÃ³Å¼nych powierzchni na cenÄ™")
plt.xlabel("Powierzchnia (stopy kw.)")
plt.ylabel("Cena domu")
plt.show()

print("""
Wnioski:
1. Zmienna docelowa to 'SalePrice' â€“ cena sprzedaÅ¼y domu.
2. RozkÅ‚ad cen jest prawoskoÅ›ny â€“ wiÄ™kszoÅ›Ä‡ domÃ³w jest w niÅ¼szej pÃ³Å‚ce cenowej.
3. WystÄ™pujÄ… outliery â€“ bardzo drogie domy (mogÄ… to byÄ‡ luksusowe posiadÅ‚oÅ›ci).
4. Powierzchnia mieszkalna (GrLivArea) ma silnÄ… korelacjÄ™ z cenÄ….
5. Powierzchnia dziaÅ‚ki (LotArea) ma sÅ‚abszÄ… zaleÅ¼noÅ›Ä‡ â€“ duÅ¼e dziaÅ‚ki nie zawsze oznaczajÄ… wysokÄ… cenÄ™.
""")
```
